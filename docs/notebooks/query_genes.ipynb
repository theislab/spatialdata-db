Instead of filtering for sample-specific metadata, we can query for datasets that contain the expression of specific genes, e.g the expression of `Gfap` or you can query using a specific `EnsemblID`:

```python
symbol_query = ln.Artifact.filter(feature_sets__genes__symbol="Gfap")
symbol_query.df(include=['otype'], features=True).head(5)
```

An example query for `ENSMUSG00000061718`. Since this gene is specific for mouse (`ENSMUSG`), the returned datasets will all be `mouse` datasets.

```python
uid_query = ln.Artifact.filter(feature_sets__genes__ensembl_gene_id="ENSMUSG00000061718")
uid_query.df(include=['otype'], features=True).head(5)
```

Download one of the selected datasets with `artifact.path.download_to(local_path: str)`.

**Known issues:**
At the moment, the download function does not check whether there's already a dataset stored at the target_path. If downloaded again to the same path, a new `spatialdata` object will be created within the first `zarr` object.

```python
artifact = ln.Artifact.get('1liSHq6aonoPkcMu0000')
target_path = './artifact_with_ENSMUSG00000061718.zarr'
artifact.path.download_to(target_path)
```

```python
sdata = sd.read_zarr(target_path)
sdata
```

```python
(
    sdata
    .pl.render_images(elements="V1_Mouse_Brain_Sagittal_Anterior_Section_2_hires_image")
    .pl.render_shapes(elements="V1_Mouse_Brain_Sagittal_Anterior_Section_2", color="ENSMUSG00000061718")
    .pl.show("downscaled_hires")
)
```
