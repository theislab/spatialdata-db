```python
import lamindb as ln
import shutil
import spatialdata as sd
from spatialdata_db.integrations import store_dataset
```

Get a list of available datasets:

```python
ln.Artifact.df()
```

Select an example uid - e.g. `wlgPi3jjfVoYUWjA0000` and retrieve the artifact:

```python
artifact = ln.Artifact.get("wlgPi3jjfVoYUWjA0000")
```

`store_dataset` will download the data associated to the artifact. Default path is `.` and the default name is the artifact uid. A different path and name can be used via the parameters `path` and `name`.

```python
path = store_dataset(artifact=artifact)
```

```python
sdata = sd.read_zarr(path)
```

```python
shutil.rmtree(path)
```
